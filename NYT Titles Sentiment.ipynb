```python
import pandas as pd
df = pd.read_csv("NYT_Articles/ArticlesApril2017.csv")
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
text = " ".join(title for title in df.headline)
text = text.lower()
print(stop_words)
```

```python
from collections import Counter

splt = text.split()
counter = Counter(word for word in splt if word not in stop_words)

print(counter.most_common(10))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

sentiment = SentimentIntensityAnalyzer()

total_sent = sentiment.polarity_scores(word for word in text if word not in stop_words)
print(total_sent)
```
